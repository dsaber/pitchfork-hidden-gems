```python
# standard
%matplotlib inline
import pandas as pd
import numpy as np

# storage
import cPickle

# clustering
from sklearn.cluster import KMeans
```

```python
tfidf, logreg = cPickle.load(open('../tfidf_logreg.pkl', 'r'))

df = pd.read_csv('../data/final_p4k.csv')
columns_to_keep = [
    'Album',
    'Artist',
    'Content',
    'Date',
    'Reviewer',
    'Score',
    'Genre',
    'Mid?',
    'NLTK_score',
    'TB_score',
    'MY_score'
]
df = df[columns_to_keep]
df.head()
```

```python
# compute amount by which albums are 'underrated'
df['NLTK_under'] = df['NLTK_score'] - df['Score']
df['TB_under']   = df['TB_score'] - df['Score']
df['MY_under']   = df['MY_score'] - df['Score']
```

```python
print df['Reviewer'].value_counts()
```

```python
print df['Genre'].value_counts()
```

```python
df_sub = df[df['Genre'] == 'Alternative & Punk,Alternative Rock,Alternative']
df_sub = df[df['Reviewer'] == 'Joe Tangari']
```

```python
joe_alt_rock = tfidf.transform(df_sub['Content'])
```

```python
km = KMeans(20)
km.fit(joe_alt_rock)
df_sub['Cluster'] = km.predict(joe_alt_rock)
```

```python
df_inspect = df_sub[['Score', 'NLTK_under', 'MY_under']].groupby(df_sub['Cluster']).mean()
df_inspect['Counts'] = df_sub['Cluster'].groupby(df_sub['Cluster']).count()
df_inspect
```
