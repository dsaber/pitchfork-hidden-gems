```python
# standard
%matplotlib inline
import pandas as pd
import numpy as np

# storage
import cPickle

# clustering
from sklearn.cluster import KMeans
```

```python
tfidf, logreg = cPickle.load(open('../tfidf_logreg.pkl', 'r'))
```

```python
df = pd.read_csv('../data/final_p4k.csv')
columns_to_keep = [
    'Album',
    'Artist',
    'Content',
    'Date',
    'Reviewer',
    'Score',
    'Genre',
    'Mid?',
    'NLTK_score',
    'TB_score',
    'MY_score'
]
df = df[columns_to_keep]
df = df.dropna()
df.head()
```

```python
# compute amount by which albums are 'underrated' according 
# to various sentiment analysis techniques

# (1) The following code rearranges score by all values
score = df[['Score']].sort('Score').values

df = df.sort('NLTK_score')
df['NLTK_scaled'] = score.copy()

df = df.sort('TB_score')
df['TB_scaled'] = score.copy()

df = df.sort('MY_score')
df['MY_scaled'] = score.copy()
```

```python
joe_alt_rock = tfidf.transform(df_sub['Content'])
```

```python
km = KMeans(4)
km.fit(joe_alt_rock)
df_sub['Cluster'] = km.predict(joe_alt_rock)
```

```python
df_inspect = df_sub[['Score', 'NLTK_scaled', 'MY_scaled']].groupby(df_sub['Cluster']).mean()
df_inspect['Counts'] = df_sub['Cluster'].groupby(df_sub['Cluster']).count()
df_inspect
```
